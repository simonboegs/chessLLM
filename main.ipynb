```python
import chess
import json
import os
import matplotlib.pyplot as plt
import random
```

```python
filenames = os.listdir("blitz-tactics-puzzles/lichess-puzzles-v1")
ratings = []
for file in filenames:
    path = f"blitz-tactics-puzzles/lichess-puzzles-v1/{file}"
    with open(path) as f:
        data = json.load(f)
    ratings.append(data["metadata"]["rating"])

plt.hist(ratings, density=True)
```

```python
plt.hist(ratings, 50, density=True)
```

```python
filenames_sample = random.sample(filenames,150)
```

```python
def split_to_json(filenames, output_file, sort=True):
    data = []
    for filename in filenames:
        path = f"blitz-tactics-puzzles/lichess-puzzles-v1/{filename}"
        with open(path) as f:
            data.append(json.load(f))

    if sort:
        data = sorted(data, key=lambda item: item["metadata"]["rating"])

    with open(output_file,"w") as f:
        f.write(json.dumps(data))

split_to_json(filenames_sample[0:100], "dataset.json")
split_to_json(filenames_sample[100:], "dataset_few_shot.json", sort=False)
```

```python
def view_puzzle(number):
    with open(f"blitz-tactics-puzzles/lichess-puzzles-v1/{number}.json") as f:
        data = json.load(f)
    fen = data["puzzle_data"]["puzzle_fen"]
    rating = data["metadata"]["rating"]
    board = chess.Board(fen)
    print(rating)
    print(fen)
    return board

view_puzzle(31968)
```
